### Aqui eu fiquei curioso para ver os valores a média ao longos do 21 anos em cada país

```python
df_mean_prec_country = df_mean_prec.mean(axis=1)
df_mean_prec_country
```

```python
df_mean_prec_country = pd.DataFrame(df_mean_prec_country)
df_mean_prec_country.columns = ['Prec']
df_mean_prec_country
```

### Para ver como a média se comportou em toda a amazônia internacional ao longo do período analiado, retirei uma média anual de toda região e gerei um novo dataframe

```python
df_mean_prec_am = df_mean_prec.mean()
df_mean_prec_am
```

```python
df_mean_prec_am = pd.DataFrame(df_mean_prec_am)
df_mean_prec_am
```

```python
df_mean_prec_am.rename_axis('Year', inplace=True)
df_mean_prec_am.columns = ['Prec']
df_mean_prec_am
```

```python
df_mean_prec_am.info()
```

### Aqui a plotagem de um gráfico de linha para visualizar o comportamento da precipitação ao longo do tempo escolhido (21 anos)

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=df_mean_prec_am, x=df_mean_prec_am.index, y="Prec")
plt.show()
```

A partir dele foi possível perceber que: 

- A cada aproximadamente 2-3 anos ocorre um aumento do valor da precipitação ao longo da amazônia internacional, seguido por um ano de diminuição desse valor;

- Apesar desse ciclo de aumento e diminuição da média de precipitação, como um todo ao longo do período análisado a precipitação na Amazônia Internacional tem diminuído.
